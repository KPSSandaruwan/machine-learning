```
import spacy
nlp = spacy.load("en_core_web_sm")
```

```
my_string = '"We\'re moving to L.A.!"'
```

```
doc = nlp(my_string)
```

```
for token in doc:
  print(token.text)
```

```
doc2 = nlp(u"Let's visit St. Louis in the U.S. next year.")
```

```
for t in doc2:
  print(t)
```

```
len(doc2) # Number of tokens
```

```
doc3 = nlp(u'Apple to build a Hong Kong factory for $8 million')
```

```
for token in doc3:
  print(token.text, end = ' | ')
```

```
for entity in doc3.ents:
  print(entity)
  print(entity.label_)
  print(str(spacy.explain(entity.label_)))
  print('\n')
```

```
doc4 = nlp(u'Autonomous cars shift insurance liability toward manufactures.')
```

```
for chunk in doc4.noun_chunks:
  print(chunk)
```
