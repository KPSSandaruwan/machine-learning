```
df = pd.read_csv('spam.csv')
df.head()
```

```
df.drop(['Unnamed: 2', 'Unnamed: 3', 'Unnamed: 4'], axis='columns', inplace=True)
```

```
df.head()
```

```
df.groupby('v1').describe()
```

```
df['spam'] = df['v1'].apply(lambda x:1 if x== 'spam' else 0)
```

```
df.head()
```

```
from sklearn.feature_extraction.text import CountVectorizer
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.v2, df.spam, test_size=0.25)
```

```
v = CountVectorizer()
X_train_count = v.fit_transform(X_train.values)
type(X_train_count)
```

```
type(X_train_count.toarray())
```

```
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count.toarray(), y_train)
```

```
email = ['Hey Guys Whats Up with you all', 'Sale, sale. with 20% of discount']
email_count = v.transform(email)
model.predict(email_count)
```
